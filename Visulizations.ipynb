```python
# Introduction to Matplotlib
```

```python
# Line Plot
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.title('Basic Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Scatter Plot
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='red')
plt.title('Basic Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Customizing Plots
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle='--', color='blue', marker='x')
plt.title('Customized Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Adding Annotations
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.title('Line Plot with Annotation')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.annotate('Max Value', xy=(5, 11), xytext=(3, 12),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
# Subplots
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(x, y1, 'r-o')
ax1.set_title('Subplot 1')
ax2.scatter(x, y2, color='blue')
ax2.set_title('Subplot 2')

plt.show()
```

```python
# Introduction to Seaborn
```

```python
# Histogram
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 4, 5, 5, 5, 6, 7, 8]
sns.histplot(data, kde=False, bins=5)
plt.title('Histogram')
plt.show()
```

```python
# Box Plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
#  Customizing Plots
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
data = sns.load_dataset('tips')
sns.scatterplot(x='total_bill', y='tip', data=data, hue='day')
plt.title('Scatter Plot with Seaborn Aesthetics')
plt.show()
```

```python
# Adding Regression Lines
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
sns.lmplot(x='total_bill', y='tip', data=data, hue='day')
plt.title('Regression Line Plot')
plt.show()
```

```python
# Advanced Plots  Pair Plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')
sns.pairplot(data, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
# Combining Seaborn and Matplotlib
```

```python
# Overlaying Seaborn and Matplotlib Plots
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], kde=True)

plt.axvline(data['total_bill'].mean(), color='red', linestyle='--', label='Mean')
plt.legend()
plt.title('Histogram with Mean Line')
plt.show()
```

```python
# Customizing Seaborn Plots with Matplotlib
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
ax = sns.boxplot(x='day', y='total_bill', data=data)

ax.set_title('Customized Box Plot')
ax.set_xlabel('Day of Week')
ax.set_ylabel('Total Bill')
plt.show()
```
